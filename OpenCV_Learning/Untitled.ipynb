```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
ori = cv2.imread( 'F:OpenCV3/moni.jpg', 0 )
_, nw = cv2.threshold( ori, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU )
plt.imshow(ori)
plt.show()
plt.imshow(nw)
plt.show()
```
