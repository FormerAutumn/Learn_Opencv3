```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
h1 = cv2.imread( 'F:\OpenCV3/hist_1.jpg', 0 )
plt.imshow(h1)
plt.axis('off')
plt.show()

hist_h1 = cv2.calcHist( [h1], [0], None, [256], [0, 256] )
plt.plot( hist_h1 )
plt.show()
```

```python
equ_h1 = cv2.equalizeHist( h1 )
plt.imshow( np.hstack( (h1, equ_h1) ) )
plt.axis('off')
plt.show()

hist_equ = cv2.calcHist( [equ_h1], [0], None, [256], [0, 256] )
plt.subplot(1,2,1)
plt.plot( hist_h1 )

plt.subplot(1,2,2)
plt.plot( hist_equ )

plt.show()
```

```python
ad_hist = cv2.imread( 'F:\OpenCV3/ad_hist.jpg', 0 )
tmp = ad_hist
equ_ad_hist = cv2.equalizeHist( ad_hist )
ad_hist_clahe = cv2.createCLAHE( clipLimit=2.0, tileGridSize=(8,8) )
ad_hist_clahe.apply( tmp )
plt.imshow( np.hstack((equ_ad_hist,tmp)) )
plt.axis('off')
plt.show()
```
