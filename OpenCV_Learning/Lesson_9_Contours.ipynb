```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
ori = cv2.imread( 'F:\OpenCV3/contours_13.jpg' )
ori_gray = cv2.cvtColor( ori, cv2.COLOR_BGR2GRAY )
_, thres = cv2.threshold( ori_gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU )

_, contours, hierarchy = cv2.findContours(
    thres, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE )
cv2.drawContours( ori, contours, -1, (255,0,255), 2 )
plt.imshow(ori)
plt.show()
```

```python
#convex
polyDP = cv2.imread( 'F:\OpenCV3/polyDP.jpg', 0 )
_, thresh = cv2.threshold( polyDP, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU )
nw, contours, hierarchy = cv2.findContours( thresh, 3, 2 )

approx = cv2.approxPolyDP( contours[0], 3, True )
nw = cv2.cvtColor( nw, cv2.COLOR_GRAY2BGR )
cv2.polylines( nw, [approx], True, (255,0,255), 2 )
plt.imshow(nw)
plt.show()
```

```python
cex = cv2.imread( 'F:\OpenCV3/convex_tst.jpg' )
b, g, r = cv2.split(cex)
g = np.bitwise_and( g, 0 )
tmp = cv2.merge((b,g,r))
cmp = np.hstack( (cex,tmp) )
plt.imshow(cmp)
plt.axis('off')
plt.show()
```

```python
_, thresh = cv2.threshold( cv2.cvtColor(tmp, cv2.COLOR_RGB2GRAY), 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU )
nw, contours, hierarchy = cv2.findContours( thresh, 3, 2 )
ct = contours[0]
hull = cv2.convexHull( ct )
nw = cv2.cvtColor( nw, cv2.COLOR_GRAY2BGR )
cv2.polylines( nw, [hull], True, (0,0,255), 2 )
plt.imshow(nw)
plt.axis('off')
plt.show()
```
