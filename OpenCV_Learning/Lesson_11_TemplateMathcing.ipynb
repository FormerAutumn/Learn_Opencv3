```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
lena = cv2.imread( 'F:\OpenCV3/lena.jpg' )
lena_face = cv2.imread( 'F:\OpenCV3/lena_face.jpg' )
h, w = lena_face.shape[:2]
pos = cv2.matchTemplate( lena, lena_face, cv2.TM_CCOEFF )
mn_val, mx_val, mn_loc, mx_loc = cv2.minMaxLoc( pos )

names = ['mn_val', 'mx_val', 'mn_loc', 'mx_loc' ]
func = [ mn_val, mx_val, mn_loc, mx_loc ]
for i in range(len(names)):
    print(names[i], type(func[i]))
    print(func[i])

left_top = mx_loc
right_buttom = ( left_top[0] + w, left_top[1] + h )
cv2.rectangle( lena, left_top, right_buttom, (255,255,255), 2 )
plt.imshow(lena[:,:,::-1])
plt.show()
```

```python
mario = cv2.imread( 'F:\OpenCV3/mario.jpg' )
mario_gray = cv2.cvtColor( mario, cv2.COLOR_BGR2GRAY )
mario_coin = cv2.imread( 'F:\OpenCV3/mario_coin.jpg', 0 )

h, w = mario_coin.shape[:2]
tmp = cv2.matchTemplate( mario_gray, mario_coin, cv2.TM_CCOEFF_NORMED )
thresh = 0.8
loc = np.where( tmp >= thresh )
print(type(loc))

for pt in zip(*loc[::-1]):
    right_buttom = ( pt[0] + w, pt[1] + h )
    cv2.rectangle( mario, pt, right_buttom, ( 0,0,255), 2 )
plt.imshow(mario[:,:,::-1])
plt.axis('off')
plt.show()
```
