```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
img = cv2.imread('F:\OpenCV3/tree_galaxy.jpg', 1)
img = img[:,:,::-1]
plt.imshow(img)
plt.show()
```

```python
#get a pixel and amend it
px = img[100,100]
print(px)
px_blue = img[100,100,0]
print(px_blue)
img[100,100] = [255,255,255]
print(img[100,100])
```

```python
#get the h,w,c with img,shape
height, width, channels = img.shape
print((height,width,channels))
#data_type
print(img.dtype)
#the total pixels h*w*c
print(img.size)
```

```python
#something about ROI
tree = img[330:, 360:700]
plt.imshow(tree[:,:,::-1])
plt.show()
```

```python
#split the channels and merge
b, g, r = cv2.split(img)
img_t = cv2.merge((b,g,r))
plt.imshow(img_t[:,:,::-1])
plt.show()

b = img[:,:,0]
g = img[:,:,1]
r = img[:,:,2]
plt.imshow( cv2.merge((b,g,r))[:,:,::-1] )
plt.show()
```

```python
#colorspace changing
img_gray = cv2.cvtColor( img, cv2.COLOR_BGR2GRAY )
plt.imshow(img_gray)
plt.show()
#the add change_model
flags = [i for i in dir(cv2) if i.startswith('COLOR_')]
print(flags)
```

```python
#follow something with specfic color(eg:blue)
blue = np.uint8( [[[255,0,0]]] )
hsv_blue = cv2.cvtColor( blue, cv2.COLOR_BGR2HSV )
print(hsv_blue)
lower_blue = np.array([100,110,110])
upper_blue = np.array([130,255,255])

hsv = cv2.cvtColor( img, cv2.COLOR_BGR2HSV )
mask = cv2.inRange( hsv, lower_blue, upper_blue )
res = cv2.bitwise_and( img, img, mask = mask )
plt.imshow(cv2.cvtColor(res,cv2.COLOR_HSV2RGB))
plt.show()
```
