```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
#sample for erode, the kernel of dilate is the same
_j = cv2.imread( 'F:\OpenCV3/absract_j.jpg' )
kernel_rect = cv2.getStructuringElement( cv2.MORPH_RECT, (5,5) )
kernel_ellipse = cv2.getStructuringElement( cv2.MORPH_ELLIPSE, (5,5) )
kernel_cross = cv2.getStructuringElement( cv2.MORPH_CROSS, (5,5) )

er_rect = cv2.erode( _j, kernel_rect )
er_elli = cv2.erode( _j, kernel_ellipse )
er_cros = cv2.erode( _j, kernel_cross )

imgs = [ _j, er_rect, er_elli, er_cros ]
titles = [ 'original', 'rect', 'ellipse', 'cross' ]
for i in range(len(imgs)):
    plt.subplot( 2, len(imgs)/2, i + 1 )
    plt.axis('off')
    plt.imshow(imgs[i])
    plt.title(titles[i])
plt.show()
```

```python
kernel_1 = cv2.getStructuringElement( cv2.MORPH_RECT, (5,5), )
w_j = cv2.imread( 'F:\OpenCV3/white_j.jpg' )
b_j = cv2.imread( 'F:\OpenCV3/black_j.jpg' )
opening = cv2.morphologyEx( w_j, cv2.MORPH_OPEN ,kernel_1 )
#erode dilate
closeing = cv2.morphologyEx( b_j, cv2.MORPH_CLOSE, kernel_1 )
#dilate erode
imgs_cal = [ w_j, opening, b_j, closeing ]
titles = [ 'original_1', 'opening', 'original_2', 'closeing' ]
for i in range( len(titles) ):
    plt.subplot( 1, len(titles), i + 1 )
    plt.imshow( imgs_cal[i] )
    plt.title( titles[i] )
plt.show()
```

```python
blt = cv2.imread( 'F:\OpenCV3/building_gradian.jpg', 0 )
kernel_2 = cv2.getStructuringElement( cv2.MORPH_RECT, (5,5) )
#dilate - erode
blt_edge = cv2.morphologyEx( blt, cv2.MORPH_GRADIENT, kernel_2 )
imgs = [ blt, blt_edge ]
titles = [ 'priginal', 'blt_edge' ]
for i in range(len(imgs)):
    plt.subplot( 1, len(imgs), i + 1 )
    plt.imshow( imgs[i] )
    plt.title( titles[i] )
plt.show()
```

```python
blt = cv2.imread( 'F:\OpenCV3/building_gradian.jpg', 0 )
kernel_2 = cv2.getStructuringElement( cv2.MORPH_RECT, (5,5) )
#tophat src - open
tophat = cv2.morphologyEx( blt, cv2.MORPH_TOPHAT, kernel_2 )
#blackhat close - src
blackhat = cv2.morphologyEx( blt, cv2.MORPH_BLACKHAT, kernel_2 )
imgs = [ blt, tophat, blackhat ]
titles = [ 'blt',' tophat', 'blackhat' ]
for i in range(len(imgs)):
    plt.subplot( 1, len(imgs), i + 1 )
    plt.imshow( imgs[i] )
    plt.title( titles[i] )
plt.show()
```
