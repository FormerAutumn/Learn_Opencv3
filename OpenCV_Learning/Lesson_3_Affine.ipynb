```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
#Geometric Transformation of Images
img4 = cv2.imread('F:\OpenCV3/rows_columns.jpg', 0)
plt.subplot(1,2,1)
plt.title('original')
plt.imshow(img4)

plt.subplot(1,2,2)
zoom = cv2.resize( img4, None, fx = 2, fy = 2, interpolation = cv2.INTER_LINEAR )
plt.title('zoom')
plt.imshow(zoom)
plt.show()
```

```python
#Affine 2*3 matrix
#[1,0,x]
#[0,1,y]
img4 = cv2.imread('F:\OpenCV3/rows_columns.jpg',0)
rows, cols = img4.shape[:2]
M = np.float32( [[1,0,100],[0,1,50]] )
nw = cv2.warpAffine( img4, M, (cols,rows) )
plt.subplot( 1, 2, 1 )
plt.title('original')
plt.imshow(img4)

plt.subplot( 1, 2, 2 )
plt.title('After affine')
plt.imshow( nw )
plt.show()
```

```python
img4 = cv2.imread('F:\OpenCV3/rows_columns.jpg',0)
rows, cols = img4.shape[:2]
M = cv2.getRotationMatrix2D( (cols/2, rows/2), 45, 0.5 )
nw = cv2.warpAffine( img4, M, (cols, rows) )
plt.subplot(2,2,1)
plt.title('original')
plt.imshow(img4)
plt.subplot(1,2,2)
plt.title('After rotate')
plt.imshow(nw)
plt.show()
```

```python
img4 = cv2.imread('F:\OpenCV3/rows_columns.jpg',0)
f1 = cv2.flip( img4, -1 ) # x & y axis
f2 = cv2.flip( img4, 0 ) # x axis
f3 = cv2.flip( img4, 1 ) # y axis
fs = [ img4, f1, f2, f3 ]
titles = [ 'original', 'par < 0', 'par = 0', 'par > 0' ]
for i in range(4):
    plt.subplot( 2, 2, i + 1 )
    plt.title( titles[i] )
    plt.imshow( fs[i] )
    plt.xticks([]), plt.yticks([])
plt.show()
```
